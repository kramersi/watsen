```python
net = unet.Unet(channels=3, n_class=3, layers=4, features_root=16, cost_kwargs=dict(class_weights=[1,1,1]))
```

## Generation of qualitative flow plots


```python
data = net.predictAll(
    model_path="./unet_trained/v8_b4_ly4_(removed bad training)/model.cpkt", 
    images_path='F:/MOY_PhD_DATA_LOCAL/2017_watsen/data/frames',
    count=2000)
```

```python
import pandas as pd
dattt = pd.DataFrame({
    'time':data['time'] , 
    'sensor':data['sensor'] , 
    'p95':data['watsen']['0.8'],
    'p95':data['watsen']['0.9'],
    'p95':data['watsen']['0.95'],
    'p97':data['watsen']['0.97']
})
# dattt.to_csv("output.csv")
```

```python
# Get training data for plotting
import glob
from datetime import datetime



training = {
    'time':[],
    'sensor':[]
}
for d in glob.glob('training_data/3_cl/*.tiff'):
    
    info = os.path.basename(d).split('_')
    time = datetime.strptime(info[0], '%y%m%d %H%M%S')
    if time > min(dattt['time']) and time < max(dattt['time']):
        training['time'].append(datetime.strptime(info[0], '%y%m%d %H%M%S'))
        training['sensor'].append(float(info[1].split('.')[0]) / 10)
        
    
```

## Plot with matplotlib


```python
fig, ax1 = plt.subplots(figsize=(16, 4))
p1=ax1.plot(dattt['time'], dattt['p95'].rolling(1, center=True).mean(), 'r', label='WATSEN')

ax2=ax1.twinx()
p2=ax2.plot(dattt['time'], dattt['sensor'], label='real water level')
p3=ax2.scatter(training['time'], training['sensor'], marker='x', s = 40, c='black', label='training data')

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
plt.legend(lines+lines2 , labels+labels2 , loc=0)

ax1.set_xlabel('time')
ax1.set_ylabel('WATSEN prediction [%]')
ax2.set_ylabel('monitored water level [cm]')

plt.show
```
