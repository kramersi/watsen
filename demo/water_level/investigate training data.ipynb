```python
import os
import glob
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
labeldir = 'training_data/3_cl/*.tiff'
```

```python
files = sorted(glob.glob(labeldir))
```

```python
fig, ax = plt.subplots(20,4, sharey=True, figsize=(18,60))

ax[0,1].set_title('Not water class')
ax[0,2].set_title('Water class')
ax[0,3].set_title('Wet')

for i in range(0,20):
    
    
    img = np.array(Image.open(files[i+0]), np.float32)
    
    
    
    
    
    ax[i,0].set_title(os.path.basename(files[i+0]))
    
    
    ax[i,1].imshow(img[...,0], aspect="auto")
    ax[i,2].imshow(img[...,1], aspect="auto")
    ax[i,3].imshow(img[...,2], aspect="auto")
```
